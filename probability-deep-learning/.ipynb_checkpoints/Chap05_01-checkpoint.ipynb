```python
import tensorflow as tf
import tensorflow_probability as tfp
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras import backend as K
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from typing import *

tfd = tfp.distributions
tfb = tfp.bijectors
```

```python
d = tfd.Normal(loc=3, scale=1.5)           #A
x = d.sample(2) # Draw two random points.    #B
px = d.prob(x) # Compute density/mass.       #C
print(x)
print(px)
```

```python
dist = tfd.Normal(loc=1.0, scale=0.1)
print('sample   :', dist.sample(3).numpy()) #Samples 3 numbers
print('prob     :',dist.prob((0,1,2)).numpy()) #Calculates the probabilities for positions 0,1,2
print('log_prob :',dist.log_prob((0,1,2)).numpy()) #Same as above just log
print('cdf      :',dist.cdf((0,1,2)).numpy()) #Calculates the cummulative distributions
print('mean     :',dist.mean().numpy()) #Returns the mean of the distribution
print('stddev   :',dist.stddev().numpy())
```

### Simulate the fishlike data

In the next few cells you will simulate some (x,y)-data where the y increases on average linerly with x but has non-constant variance that so that scatterplot looks like a fish.You will fist simulate random distributed noise with non constant variance, then uniformly distributed x values between -1 and 6 and finally calculate corresponding y values with y = 2.7*x+noise (linear slope of 2.7 and intercept of 0). The variance of the noise will change, it starts with a high value of 12 and gets smaller until it is reaches a constant value of 1, then it grows again until a value of 15 to stay constant for a while and to decrease to 1 again. Look at the plot to understand the behavior of the variance. 

```python
sigma = np.concatenate(
    [np.arange(12, 1, -0.1), np.repeat(1, 30), np.arange(1, 15, 0.1), np.repeat(15, 50), np.arange(15, 1, -0.1), np.repeat(1, 20)])
plt.plot(sigma)
plt.xlabel("index", size=16)
plt.ylabel("sigma", size=16)
plt.show()
```

```python
# generation the x values for the simulated data
np.random.seed(4710)
noise = np.random.normal(0, sigma, len(sigma))

np.random.seed(99)
first_part = 110

np.random.seed(97)
x = np.concatenate(
    [np.random.uniform(-1, 1, first_part), np.random.uniform(1, 6, len(noise) - first_part)])

x = np.sort(x)
y = 2.7 * x + noise
plt.scatter(x, y, color="steelblue") 
plt.xlabel("x", size=16)
plt.ylabel("y", size=16)
plt.plot(x, 2.7 * x, color="black")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=47)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=22)

print("nr of traning samples = ",len(x_train))
print("nr of validation samples = ",len(x_val))
print("nr of test samples = ",len(x_test))
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.scatter(x_train, y_train, color="steelblue")
plt.xlabel("x", size=16)
plt.ylabel("y", size=16)
plt.title("train data")
plt.xlim([-1.5, 6.5])
plt.ylim([-30, 55])


plt.subplot(1, 2, 2)
plt.scatter(x_val, y_val, facecolors='none', edgecolors="steelblue")
plt.xlabel("x", size=16)
plt.ylabel("y", size=16)
plt.title("validation data")
plt.xlim([-1.5, 6.5])
plt.ylim([-30, 55])
plt.show()
```

## Fit a linear regression model with constant variance

Using TFP for linear regression with constant variance

```python
def NLL(y, distr): 
  return -distr.log_prob(y) 


def generate_constant_var_model(sigma: int):
    inputs = Input(shape=(1,))
    loc = Dense(1)(inputs)
    distr = tfp.layers.DistributionLambda(lambda t: tfd.Normal(loc=t, scale=sigma))(loc)
    # mean_pred_func = K.function(inputs=inputs, outputs=loc)
    model = Model(inputs=inputs, outputs=distr, name=f"model_sigma_{sigma}")
    model.compile(Adam(), loss=NLL)
    return model
```

```python
model_sd_1 = generate_constant_var_model(sigma=1)
model_sd_1.summary()
```

```python
history = model_sd_1.fit(x_train, y_train, epochs=1000, verbose=0, validation_data=(x_val, y_val))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.ylabel('NLL')
plt.xlabel('Epochs')
plt.show()
```

```python
print(model_sd_1.evaluate(x_train, y_train, verbose=0))
print(model_sd_1.evaluate(x_val, y_val, verbose=0))
```

```python
# # define a model that predicts only the mean of the output distribution.
# mean_distr = tfp.layers.DistributionLambda(make_distribution_fn=my_distr, convert_to_tensor_fn=tfd.Distribution.mean)(params)
# model_mean_sd_1 = Model(inputs=inputs, outputs=mean_distr)  # according to the `_mean` method defined by tfp's Normal distribution, the output is the `loc` itself
```

```python
def plot_predict_band(pred_func: Callable, sigma_func: Callable):
    x_pred = np.arange(-1, 6, 0.1)
    preds = pred_func(x_pred)
    sigma = sigma_func(x_pred)
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))
    xy = [(x_train, y_train), (x_val, y_val)]
    labels = ["train data", "validation data"]
    for i, ((x, y), label) in enumerate(zip(xy, labels)):
        ax = axes[i]
        ax.scatter(x, y, color="steelblue")
        ax.plot(x_pred, preds, color="black", linewidth=2)
        ax.plot(x_pred, preds + 2 * sigma, color="red", linestyle="--", linewidth=2)
        ax.plot(x_pred, preds - 2 * sigma, color="red", linestyle="--", linewidth=2)
        ax.set_xlabel("x", size=16)
        ax.set_ylabel("y", size=16)
        ax.set_title(label)
        ax.set_xlim(-1.6, 6.5)
        ax.set_ylim(-30, 55)
    plt.show()
```

```python
plot_predict_band(lambda t: model_sd_1(t).mean(), lambda _: 1)
```

```python
model_sd_5 = generate_constant_var_model(sigma=5)
model_sd_5.summary()
```

```python
history = model_sd_5.fit(x_train, y_train, epochs=1000, verbose=0, validation_data=(x_val, y_val))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.ylabel('NLL')
plt.xlabel('Epochs')
plt.show()
```

```python
print(model_sd_5.evaluate(x_train, y_train, verbose=0))
print(model_sd_5.evaluate(x_val, y_val, verbose=0))
```

```python
plot_predict_band(lambda t: model_sd_5(t).mean(), lambda _: 5)
```

## Fit a linear regression model and allow the sd to depend in a monotonic way on the input

```python
def NLL(y, distr): 
  return -distr.log_prob(y) 


def generate_monotoic_var_model():
    inputs = Input(shape=(1,))
    params = Dense(2)(inputs)
    distr = tfp.layers.DistributionLambda(lambda t: tfd.Normal(loc=t[:, :1], scale=1e-3 + K.softplus(0.05 * t[:,1:2])))(params)
    # mean_pred_func = K.function(inputs=inputs, outputs=loc)
    model = Model(inputs=inputs, outputs=distr, name="model_monotoic_sigma")
    model.compile(Adam(learning_rate=0.01), loss=NLL)
    return model
```

```python
model_monotoic_sd = generate_monotoic_var_model()
model_monotoic_sd.summary()
```

```python
history = model_monotoic_sd.fit(x_train, y_train, epochs=6000, verbose=0, validation_data=(x_val, y_val))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.ylabel('NLL')
plt.xlabel('Epochs')
plt.show()
```

```python
print(model_monotoic_sd.evaluate(x_train, y_train, verbose=0))
print(model_monotoic_sd.evaluate(x_val, y_val, verbose=0))
```

```python
plot_predict_band(lambda t: model_monotoic_sd(t).mean(), lambda t: model_monotoic_sd(t).stddev())
```

## Fit a linear regression and allow the sd to depend in a flexible manner on the input

```python
def NLL(y, distr): 
  return -distr.log_prob(y) 


def generate_flexible_var_model():
    inputs = Input(shape=(1,))
    loc = Dense(1)(inputs)
    h = Dense(30, activation="relu")(inputs)
    h = Dense(20, activation="relu")(h)
    h = Dense(20, activation="relu")(h)
    std = Dense(1)(h)
    params = Concatenate(axis=1)([loc, std])
    distr = tfp.layers.DistributionLambda(lambda t: tfd.Normal(loc=t[:, :1], scale=1e-3 + K.softplus(0.05 * t[:,1:2])))(params)
    # mean_pred_func = K.function(inputs=inputs, outputs=loc)
    model = Model(inputs=inputs, outputs=distr, name="model_flexible_sigma")
    model.compile(Adam(learning_rate=0.01), loss=NLL)
    return model
```

```python
model_flex_sd = generate_flexible_var_model()
model_flex_sd.summary()
```

```python
history = model_flex_sd.fit(x_train, y_train, epochs=2000, verbose=0, validation_data=(x_val, y_val))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.ylabel('NLL')
plt.xlabel('Epochs')
plt.show()
```

```python
print(model_flex_sd.evaluate(x_train, y_train, verbose=0))
print(model_flex_sd.evaluate(x_val, y_val, verbose=0))
```

```python
plot_predict_band(lambda t: model_flex_sd(t).mean(), lambda t: model_flex_sd(t).stddev())
```
